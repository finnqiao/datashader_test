```python
# Parse out non Californian businesses.
df = df[df['City'] == 'San Francisco']
```

```python
# Only keep locations that have latitude and longitude values.
df = df.dropna(subset = ['Business Location'])
```

```python
# Only keep active locations.
df = df[df['Business End Date'].isna()]
```

```python
lat_re = re.compile("(?<= \()\-?\d+\.\d+(?=\,)")
lon_re = re.compile("(?<=\, )\-?\d+\.\d+(?=\))")

test = re.compile('.+', re.MULTILINE)

df['Business Location'] = df['Business Location'].str.replace('\n',' ')
df['Business Location'][6]
```

```python
float(lat_re.findall(df['Business Location'][6])[0])
```

```python
# lat_re.findall(df.loc[13,]['Business Location']) == []
```

```python
# Functions to get latitude and longitude from the appropriate column with each row in the dataframe.
def get_lat(row):
    if lat_re.findall(row['Business Location']) == []:
        return np.NAN
    else: 
        return float(lat_re.findall(row['Business Location'])[0])
    
def get_lon(row):
    if lon_re.findall(row['Business Location']) == []:
        return np.NAN
    else: 
        return float(lon_re.findall(row['Business Location'])[0])
```

```python
df['lat'] = df.apply(get_lat, axis=1)
df['lon'] = df.apply(get_lon, axis=1)
```

```python
df = df.dropna(subset = ['lat','lon'])
```

```python
df.loc[:, 'easting'], df.loc[:, 'northing'] = webm(df['lon'],df['lat'])
```

```python
# Only keep relevant columns for plotting.
df = df[['DBA Name','Business Start Date', 'Location Start Date','NAICS Code Description','lat','lon','easting','northing']]
```

```python
df['Business Start Date'] = pd.to_datetime(df['Business Start Date'])
df['Location Start Date'] = pd.to_datetime(df['Location Start Date'])
```

```python
# Age based on date in row.
date_now = datetime.now()
def get_age(row, column_str):
    return (date_now.year - row[column_str].year) * 12 + date_now.month - row[column_str].month
```

```python
# Create columns for business age and how long location has been open.
df['business_age'] = df.apply(lambda row: get_age(row, 'Business Start Date'), axis = 1)
df['location_age'] = df.apply(lambda row: get_age(row, 'Location Start Date'), axis = 1)
```

```python
df.head()
```

```python
# Getting range of latitude and longitude for plotting later.
y_range_min = df['lat'].quantile(0.01)
y_range_max = df['lat'].quantile(0.99)
x_range_min = df['lon'].quantile(0.01)
x_range_max = df['lon'].quantile(0.99)
```

```python
# Creating zip object with ranges for SF to plot.
sw = webm(x_range_min,y_range_min)
ne = webm(x_range_max,y_range_max)
SF = zip(sw, ne)
```

```python
dask_df = dd.from_pandas(df, npartitions=mp.cpu_count())
```

```python
dask_df = dask_df.compute()
```
